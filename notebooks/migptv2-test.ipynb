```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Update this path if your model is mounted elsewhere
model_path = "/kaggle/input/mini-instructgpt-ppo-v2-checkpoint/ppo_model_v2"

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

tok = AutoTokenizer.from_pretrained(model_path, local_files_only=True)
tok.pad_token = tok.eos_token

model = AutoModelForCausalLM.from_pretrained(model_path, local_files_only=True).to(device)
model.eval()

print("Type 'exit' to stop.")
while True:
    user = input("\nYou: ").strip()
    if user.lower() == "exit":
        break

    prompt = f"Human: {user}\n\nAssistant:"
    x = tok(prompt, return_tensors="pt").to(device)

    with torch.no_grad():
        y = model.generate(
            **x,
            max_new_tokens=120,
            do_sample=True,
            temperature=0.7,
            repetition_penalty=1.2,
            pad_token_id=tok.eos_token_id
        )

    reply = tok.decode(y[0][x["input_ids"].shape[1]:], skip_special_tokens=True)
    print("Model:", reply)
```
